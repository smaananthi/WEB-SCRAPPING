##### - exploring one of the TABLE

```python
rows = df.find('table', class_='infobox ib-region-symbols').find_all('tr')
rows
# by mentioning a class you can get a view of a specific table
```

```python
#extract text from rows
for i in rows:
  header= i.find('th',class_="infobox-label")
  data = i.find('td',class_="infobox-data")

  if header and data:
    print(header.get_text(strip=True), ':', data.get_text(strip=True))
    # strip=True ===> used to ensure that the extracted text is cleaner
    # and doesn't contain unnecessary whitespace.
    # try to remove "strip=True" and run the code
    # you can see unwanted text in output
```
